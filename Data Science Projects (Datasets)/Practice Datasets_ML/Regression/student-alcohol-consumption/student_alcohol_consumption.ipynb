```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
#plt.rcParams['figure.figsize'] = (15.0,7.0)
```

```python
df1 = pd.read_csv('student-mat.csv',sep=',')
print(df1.shape)
df1.head()
```

```python
df2 = pd.read_csv('student-por.csv',sep=',')
print(df2.shape)
df2.head()
```

```python
df1.columns
```

```python
student = pd.concat([df1,df2],ignore_index=True)
print(student.shape)
student.head()
```

```python
sns.catplot(x="sex", kind="count",palette="magma", data=student, height = 6)
plt.title("Gender of students : F - female,M - male")
```
