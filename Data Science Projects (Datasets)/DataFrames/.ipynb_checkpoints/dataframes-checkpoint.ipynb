```python
d = {'Cost': [5000, 7000, 9000, 12000],'DiscoutRate':[.1, .2,.15,.25]}
d
```

```python
df2 = pd.DataFrame(d)
df2
```

```python
df2['Cost_with_GST_18'] = df2.apply(lambda row1: row1.Cost+0.18*row1.Cost,axis =1)
df2
```

```python
df2['Cost with discount'] = df2.apply(lambda row1: row1.Cost_with_GST_18 - row1.DiscoutRate*row1.Cost_with_GST_18,axis =1)
df2
```

## Make 3 Dataframes with different values

```python
df3 = pd.DataFrame(np.array([[1, 2, 3, 4], [4, 5, 6, 7], [7, 8, 9, 0], [3, 5, 7, 9]]), index=[0, 1, 2, 3], columns=['a', 'b', 'c', 'd'])
df3
```

```python
df4 = pd.DataFrame(np.array([[0, 1, 5, 2], [7, 2, 4, 1], [4, 0, 1, 2], [2, 3, 4, 4]]), index=[4, 5, 6, 7],columns=['w', 'x', 'y', 'z'])
df4
```

```python
df5 = pd.DataFrame(np.array([[1, 3, 5, 7], [0, 2, 4, 6], [1, 5, 7, 0], [1, 3, 3, 1]]), index=[8, 9, 10, 11],columns=['a', 'b', 'c', 'd'])
df5
```

```python
df6 = pd.concat([df3,df4,df5],axis = 1) # default axis is 0
df6
```

```python
# should have common column names
df7 = pd.merge(df3,df5,on = 'b', how = 'left')
df7
```

```python
# should have different column names
df8 = df3.join(df4, how = 'outer')
df8
```
