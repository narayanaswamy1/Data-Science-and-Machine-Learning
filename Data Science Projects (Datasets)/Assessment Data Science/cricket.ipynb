```python
Team = list(ipl17['Team'])
points = list(ipl17['Points'])
win_per = list(ipl17['Win_Percent'])
dfx = pd.DataFrame(points,index=Team,columns=['Points'])
dfy = pd.DataFrame(win_per,index=Team,columns=['Win_Percent'])
dfxy = dfx.transpose().append(dfy.transpose())
dfxy
```

```python
Team = list(ipl18['Team'])
points = list(ipl18['Points'])
win_per = list(ipl18['Win_Percent'])
dfi = pd.DataFrame(points,index=Team,columns=['Points'])
dfj = pd.DataFrame(win_per,index=Team,columns=['Win_Percent'])
dfij = dfi.transpose().append(dfj.transpose())
dfij
```

```python
# Combining both tables Stats of Points and Win Percentage
combined_stats = dfxy.transpose().add(dfij.transpose(),fill_value=0)
combined_stats
```

```python
# Sorting Combimed Tables based on Points
combined_stats.sort_values('Points', ascending=False)
```
