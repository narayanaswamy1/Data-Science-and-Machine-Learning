```python
import warnings
warnings.filterwarnings('ignore')
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = (15.0,7.0)
```

```python
weather = pd.read_csv('weatherHistory.csv')
print(weather.shape)
weather.head()
```

```python
sns.heatmap(weather.corr(),annot=True,cmap='Blues_r')
```

```python
X = weather['Humidity'].values
y = weather['Temperature (C)'].values
```

```python
plt.scatter(X,y)
```

```python
X = X.reshape(len(X),1)
```

```python
print(type(X),type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
lm = lm.fit(X_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
y_pred_train = lm.predict(X_train)
y_pred_test = lm.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.r2_score(y_train,y_pred_train)
```

```python
metrics.r2_score(y_test,y_pred_test)
```

```python
plt.scatter(X, y,c='orange',label='Scatter Plot')
plt.plot(X, (-30.93932921)*X + 34.712855403365694,c='g',label='Regression Line')
plt.xlabel('Humidity')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
plt.scatter(X_test,y_test)
plt.scatter(X_test, y_pred_test, color='r')
```

```python
import statsmodels.api as sm
```

```python
X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train,X_train_sm).fit()
```

```python
lr.params
```

```python
print(lr.summary())
```
