```python
X = USA_Housing.drop(['Price', 'Address'],axis=1)
y = USA_Housing['Price'] #use double [[]] for Dataframe
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Creating and Training a model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm
```

```python
lm.fit(X_train, y_train)
```

## Model Evaluation

Let's evaluate the model by checking out its coefficients and how we can interpret them

```python
# print the intercept
print(lm.intercept_)
```

```python
lm.coef_
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

## Predictions from model

```python
y_pred = lm.predict(x_test)
```

```python
y_pred_train = lm.predict(X_train)
```

```python
len(y_pred)
```

```python
len(y_test)
```

# Evaluation

```python
from sklearn import metrics
```

```python
metrics.r2_score(y_test, y_pred)
```

```python
metrics.r2_score(y_train,y_pred_train)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.scatter(y_test, y_pred)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.scatter(y_train, y_pred_train)
```

# Residual Histogram

```python
sns.distplot((y_test-y_pred),bins=50)
```

```python
sns.distplot((y_pred),bins=50)
```

```python
sns.distplot((y_train-y_pred_train),bins=50)
```

```python
sns.distplot((y_pred_train),bins=50)
```

# Regression evaluation metrics

```python
print('MAE', metrics.mean_absolute_error(y_test,y_pred))
print('MSE', metrics.mean_squared_error(y_test,y_pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
