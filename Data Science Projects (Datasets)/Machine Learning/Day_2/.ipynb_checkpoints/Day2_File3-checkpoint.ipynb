```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
housing = pd.read_csv('Housing.csv')
housing
```

```python
housing.head(2)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
housing.mainroad=le.fit_transform(housing['mainroad'])
housing.guestroom=le.fit_transform(housing['guestroom'])
housing.basement=le.fit_transform(housing['basement'])
housing.hotwaterheating=le.fit_transform(housing['hotwaterheating'])
housing.airconditioning=le.fit_transform(housing['airconditioning'])
housing.prefarea=le.fit_transform(housing['prefarea'])
#housing.furnishingstatus=le.fit_transform(housing['furnishingstatus'])
```

```python
housing.head(2)
```

```python
status = pd.get_dummies(housing['furnishingstatus'])
status.head(3)
```

```python
# type of furnishing can be identified with just last two columns
#00 will correspond to furnished
#01 will corresspond to unfurnished
#10 will correspond to semi-furnished
```

```python
status = pd.get_dummies(housing['furnishingstatus'], drop_first=True)
status.head(5)
```

```python
housing = pd.concat([housing,status], axis = 1)
```

```python
housing.head(10)
```

```python
housing.drop(labels='furnishingstatus', axis = 1, inplace=True)
```

```python
housing.head(10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
sns.heatmap(housing.corr(),annot=True)
```

```python
housing.columns
```

```python
x = housing.loc[:,['area', 'bathrooms']]
y = housing.price
```

```python
print(type(x),type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=.3, random_state=101)
```

```python
import statsmodels.api as sm
```

```python
x_train_sm = sm.add_constant(x_train)
```

```python
lr = sm.OLS(y_train,x_train_sm).fit()
```

```python
lr.params
```

```python
print(lr.summary())
```

```python
# VIF check
#variance inflation factor
# we generally want vif that is less than 5 so there are clearly some variables we need to drop
```

```python
# Variance inflation factor or VIF, gives a basic quantitative idea about how much the feature variables are corealated 
# with each other
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = x_train.columns
vif['VIF'] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
vif
```
