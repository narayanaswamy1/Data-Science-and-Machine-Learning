```python
col = list(popularity.columns[(popularity.skew()>1.96) | (popularity.skew()<-1.96)])
col
```

```python
popularity.skew().sort_values()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = popularity
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
popularity_data = []
k = list(range(1,10)) #this line or below line
for i in range(1,10):
    km = KMeans(i).fit(X_scaled)
    y_km_popularity_pred = km.fit_predict(X_scaled)
    popularity_data.append(km.inertia_)
popularity_data
```

```python
plt.figure(figsize=(10,5))
plt.plot(k,popularity_data)
```

```python
kmeans = KMeans(n_clusters=5,init='k-means++',random_state=42)
y_means = kmeans.fit_predict(X_scaled)
y_means
```

```python
plt.scatter(X_scaled[y_means==0,1], X_scaled[y_means==0,2], s=50, c='red', label = 'Cluster 1')
plt.scatter(X_scaled[y_means==1,1], X_scaled[y_means==1,2], s=50, c='blue', label = 'Cluster 2')
plt.scatter(X_scaled[y_means==2,1], X_scaled[y_means==2,2], s=50, c='green', label = 'Cluster 3')
plt.scatter(X_scaled[y_means==3,1], X_scaled[y_means==3,2], s=50, c='grey', label = 'Cluster 4')
plt.scatter(X_scaled[y_means==4,1], X_scaled[y_means==4,2], s=50, c='yellow', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=500, c='Black', label='Centroid',marker='*')
plt.title('Popularity')
plt.xlabel('Annual Income(k$) ')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
