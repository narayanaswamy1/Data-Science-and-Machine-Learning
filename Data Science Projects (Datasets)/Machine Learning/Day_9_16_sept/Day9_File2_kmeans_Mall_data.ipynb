```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = mall.drop(['CustomerID'],axis=1)
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
mall_data = []
k = list(range(1,7)) #this line or below line
for i in range(1,7):
    km = KMeans(i).fit(X_scaled)
    y_km_mall_pred = km.fit_predict(X_scaled)
    mall_data.append(km.inertia_)
mall_data      
```

```python
plt.figure(figsize=(10,5))
plt.plot(k,mall_data)
```

```python
kmeans = KMeans(n_clusters=2,init='k-means++',random_state=42)
y_means = kmeans.fit_predict(X_scaled)
y_means
```

```python
plt.scatter(X_scaled[y_means==0,1], X_scaled[y_means==0,2], s=50, c='red', label = 'Cluster 1')
plt.scatter(X_scaled[y_means==1,1], X_scaled[y_means==1,2], s=50, c='blue', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=500, c='Black', label='Centroid',marker='*')
plt.title('Mall Customers')
plt.xlabel('Annual Income(k$) ')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
