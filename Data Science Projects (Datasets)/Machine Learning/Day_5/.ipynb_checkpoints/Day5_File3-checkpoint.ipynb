```python
house_sqft_lot = house[(house['sqft_lot']>(np.mean(house['sqft_lot'])-3*np.std(house['sqft_lot']))) & (house['sqft_lot']<(np.mean(house['sqft_lot'])+.5*np.std(house['sqft_lot'])))]
print(house_sqft_lot.shape)
plt.figure(figsize=(15,5))
sns.boxplot(house_sqft_lot)
```

```python
house_sqft_lot15 = house_sqft_lot[(house_sqft_lot['sqft_lot15']>(np.mean(house_sqft_lot['sqft_lot15'])-3*np.std(house_sqft_lot['sqft_lot15']))) & (house_sqft_lot['sqft_lot15']<(np.mean(house_sqft_lot['sqft_lot15'])+2.15*np.std(house_sqft_lot['sqft_lot15'])))]
print(house_sqft_lot15.shape)
plt.figure(figsize=(15,5))
sns.boxplot(house_sqft_lot15)
```

```python
house_price = house_sqft_lot15[(house_sqft_lot15['price']>(np.mean(house_sqft_lot15['price'])-3*np.std(house_sqft_lot15['price']))) & (house_sqft_lot15['price']<(np.mean(house_sqft_lot15['price'])+2.15*np.std(house_sqft_lot15['price'])))]
print(house_price.shape)
plt.figure(figsize=(15,5))
sns.boxplot(house_price)
```

```python
house_bedrooms = house_price[(house_price['bedrooms']>(np.mean(house_price['bedrooms'])-3*np.std(house_price['bedrooms']))) & (house_price['bedrooms']<(np.mean(house_price['bedrooms'])+2*np.std(house_price['bedrooms'])))]
print(house_bedrooms.shape)
plt.figure(figsize=(15,5))
sns.boxplot(house_bedrooms)
```

```python
house_bedrooms.skew().sort_values(ascending=False)
```

```python
fig = plt.figure()
sns.heatmap(house_bedrooms.corr(),annot=True,annot_kws={'size':13},ax=fig.add_axes([0,0,3,3]))
```

```python
from sklearn.model_selection import train_test_split
```

```python
house_bedrooms_train, house_bedrooms_test = train_test_split(house_bedrooms,test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
house_bedrooms_train_scaled = pd.DataFrame(sc.fit_transform(house_bedrooms_train),columns=house_bedrooms.columns)
```

```python
house_bedrooms_test_scaled = pd.DataFrame(sc.fit_transform(house_bedrooms_test),columns=house_bedrooms.columns)
```

```python
print(house_bedrooms_train_scaled.shape)
```

```python
print(house_bedrooms_test_scaled.shape)
```

```python
X_train = house_bedrooms_train_scaled.drop('price',axis=1)
X_test = house_bedrooms_test_scaled.drop('price',axis=1)
```

```python
y_train = house_bedrooms_train_scaled.price
y_test = house_bedrooms_test_scaled.price
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
rfe = RFE(lr,10)
```

```python
rfe = rfe.fit(X_train,y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
rfe.support_
```

```python
rfe.ranking_
```

```python
selected_col = X_train.columns[rfe.support_]
```

```python
selected_col
```

```python
X_train_rfe = X_train[selected_col]
```

```python
X_train_rfe.head(5)
```

```python
import statsmodels.api as sm
```

```python
X_train_rfe_sm = sm.add_constant(X_train_rfe)
```

```python
linmod = sm.OLS(y_train,X_train_rfe_sm).fit()
```

```python
print(linmod.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe_sm.values, i) for i in range(X_train_rfe_sm.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
print(vif)
```

```python
X_train_rfe_sm.drop('sqft_basement',axis=1,inplace=True)
```

```python
linmod1 = sm.OLS(y_train,X_train_rfe_sm).fit()
print(linmod1.summary())

vif = pd.DataFrame()
vif['Features'] = X_train_rfe_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe_sm.values, i) for i in range(X_train_rfe_sm.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
print(vif)
```

```python
y_train_sm_pred = linmod1.predict(X_train_rfe_sm)
```

```python
error = y_train-y_train_sm_pred
```

```python
sns.distplot(error)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_sm_pred)
```

```python
X_test_sm = sm.add_constant(X_test)
```

```python
X_test_sm = X_test_sm[X_train_rfe_sm.columns]
```

```python
y_pred_test_sm = linmod1.predict(X_test_sm)
y_pred_train_sm = linmod1.predict(X_train_rfe_sm)
```

```python
r2_score(y_test,y_pred_test_sm)
```

```python
lr.fit(X_train_rfe_sm,y_train)
```

```python
y_pred_Lr = lr.predict(X_test_sm)
```

```python
r2_score(y_test,y_pred_Lr)
```
