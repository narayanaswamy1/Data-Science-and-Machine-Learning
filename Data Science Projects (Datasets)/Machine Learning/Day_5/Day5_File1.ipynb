```python
wine_rs = wine[(wine['residual sugar']>(np.mean(wine['residual sugar'])-3*np.std(wine['residual sugar']))) & (wine['residual sugar']<(np.mean(wine['residual sugar'])+2.15*np.std(wine['residual sugar'])))]
print(wine_rs.shape)
plt.figure(figsize=(15,5))
sns.boxplot(wine_rs)
```

```python
wine_cl = wine_rs[(wine_rs['chlorides']>(np.mean(wine_rs['chlorides'])-3*np.std(wine_rs['chlorides']))) & (wine_rs['chlorides']<(np.mean(wine_rs['chlorides'])+2.7*np.std(wine_rs['chlorides'])))]
print(wine_cl.shape)
plt.figure(figsize=(15,5))
sns.boxplot(wine_cl)
```

```python
wine_sp = wine_cl[(wine_cl['sulphates']>(np.mean(wine_cl['sulphates'])-3*np.std(wine_cl['sulphates']))) & (wine_cl['sulphates']<(np.mean(wine_cl['sulphates'])+3*np.std(wine_cl['sulphates'])))]
print(wine_sp.shape)
plt.figure(figsize=(15,5))
sns.boxplot(wine_sp)
```

```python
from sklearn.model_selection import train_test_split
```

```python
wine_sp_train, wine_sp_test = train_test_split(wine_sp,test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
wine_sp_train_scaled = pd.DataFrame(sc.fit_transform(wine_sp_train),columns=wine_sp.columns)
```

```python
wine_sp_test_scaled = pd.DataFrame(sc.fit_transform(wine_sp_test),columns=wine_sp.columns)
```

```python
print(wine_sp_train_scaled.shape)
```

```python
print(wine_sp_test_scaled.shape)
```

```python
X_train = wine_sp_train_scaled.drop('pH',axis=1)
X_test = wine_sp_test_scaled.drop('pH',axis=1)
```

```python
y_train = wine_sp_train_scaled.pH
y_test = wine_sp_test_scaled.pH
```

```python
from sklearn.feature_selection import RFE
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
rfe = RFE(lr,9)
```

```python
rfe = rfe.fit(X_train,y_train)
```

```python
X_train_rfe = X_train[selected_col]
```

# Statsmodel

```python
import statsmodels.api as sm
```

```python
X_train_rfe_sm = sm.add_constant(X_train_rfe)
```

```python
linmod = sm.OLS(y_train,X_train_rfe_sm).fit()
```

```python
print(linmod.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['Features'] = X_train_rfe_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe_sm.values, i) for i in range(X_train_rfe_sm.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
print(vif)
```

```python
X_train_rfe_sm.drop('citric acid',axis=1,inplace=True)
```

```python
X_train_rfe_sm.columns
```

```python
linmod1 = sm.OLS(y_train,X_train_rfe_sm).fit()
print(linmod1.summary())

vif = pd.DataFrame()
vif['Features'] = X_train_rfe_sm.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe_sm.values, i) for i in range(X_train_rfe_sm.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by = 'VIF', ascending = False)
print(vif)
```

```python
y_train_sm_pred = linmod1.predict(X_train_rfe_sm)
```

```python
error = y_train-y_train_sm_pred
```

```python
sns.distplot(error)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_sm_pred)
```
