```python
ad_trimmed = pd.read_csv('ad_trim',na_values=['?'])
ad_trimmed[ad_trimmed.index==9]
```

```python
ad_trimmed.columns = [x.strip() for x in ad_trimmed.columns]
ad_trimmed
```

```python
col_mean_fill = list(ad_trimmed.columns[0:4])
```

```python
for i in col_mean_fill:
    ad_trimmed[i].fillna(value=ad_trimmed[i].mean(),inplace=True)
```

```python
ad_trimmed.isnull().sum().head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
ad_trimmed['ad.'] = le.fit_transform(ad_trimmed['ad.'])
```

```python
ad_trimmed.head()
```

```python
set(ad_trimmed['0'])
```

```python
X = ad_trimmed.drop(['ad.'],axis=1).values
y = ad_trimmed['ad.'].values
```

```python
print(type(X),type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=20)
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

## Import calssification model

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lm = LogisticRegression()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
lm = lm.fit(X_train_pca,y_train)
```

## Train Data

```python
y_pred_lm_train = lm.predict(X_train_pca)
```

```python
print('Accuracy_Score: ',accuracy_score(y_train,y_pred_lm_train))
```

## Test Data

```python
y_pred_lm_test = lm.predict(X_test_pca)
```

```python
print('Accuracy_Score: ',accuracy_score(y_test,y_pred_lm_test))
```

```python
pca.explained_variance_ratio_
```

```python
(pca.explained_variance_ratio_*100).sum()
```

```python
pc = list(range(1,21))
plt.rcParams['figure.figsize'] = (10.0,5.0)
plt.bar(pc,pca.explained_variance_ratio_*100)
```

## Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred_test = knn.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred_test))
print()
print(confusion_matrix(y_test,y_pred_test))
print()
print(classification_report(y_test,y_pred_test))
```

```python
X = ad_trimmed.drop(['ad.'],axis=1).values
y = ad_trimmed['ad.'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
pca = PCA(n_components=40)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
lm = lm.fit(X_train_pca,y_train)
y_pred_lm_train = lm.predict(X_train_pca)
print('Accuracy_Score_Train: ',round(accuracy_score(y_train,y_pred_lm_train)*100,4))
y_pred_lm_test = lm.predict(X_test_pca)
print('Accuracy_Score_Test: ',round(accuracy_score(y_test,y_pred_lm_test)*100,4))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
y_pred_test = knn.predict(X_test)
print(accuracy_score(y_test,y_pred_test))
print()
print(confusion_matrix(y_test,y_pred_test))
print()
print(classification_report(y_test,y_pred_test))
```
